```python
import numpy as np
import matplotlib.pyplot as plt
```

# 2) definição da função Sigmoiode e sua derivada 

A Função sigmoid é dada por:

\begin{equation}
\sigma(x) = \frac{1}{ 1 + e^{-x}}
\end{equation}

e a sua deriva é,

\begin{equation}
\sigma(x) = \sigma(x) (1 - \sigma(x)) = \frac{1}{ 1 + e^{-x}} \left(1 - \frac{1}{ 1 + e^{-x}} \right)
\end{equation}


```python
def sigmod(x):
  return 1.0/(1.0 + np.exp(-x))

def dsigmod_dx(x, sigmod=sigmod):
  return sigmod(x)*(1-sigmod(x))
```

```python
x = np.linspace(start=-10, stop=10, num=100)
```

```python
s = sigmod(x)
ds_dx = dsigmod_dx(x)
```

## 2.1) Plotanto a sigmod e sua deriva

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, s, label ='sigmod')
ax.plot(x, ds_dx, label ='derivada')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)

ax.set_xlabel('x')
ax.set_xlabel('y')

ax.set_ylim(0, 1.1)
ax.set_xlim(min(x), max(x))

plt.grid(ls = '--', c = 'silver')

plt.legend(fontsize=14)

plt.show()
```
